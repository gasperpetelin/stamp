```python
from sktime.classification.hybrid import HIVECOTEV1
# from sktime.classification.shapeled import  ShapeletTransformClassifier
from sktime.classification.distance_based import  KNeighborsTimeSeriesClassifier

from sktime.utils.data_io import load_from_tsfile_to_dataframe
import os 
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold
import sktime
```

```python
path = "./Meta_learning_Time_series_Classification/original_data/Univariate2018_ts/Univariate_ts/"


# dataset_name = data.iloc[0, 0]
def knn_time_series(dataset_name, random_state):

    print("$$$$$$$"*10)
    print("Processing ", dataset_name)


    train_data_name = path + dataset_name + "/" + dataset_name + "_TRAIN.ts"
    test_data_name = path  + dataset_name + "/" +  dataset_name + "_TEST.ts"


    train_x, train_y = load_from_tsfile_to_dataframe(train_data_name, return_separate_X_and_y=True)
    test_x, test_y = load_from_tsfile_to_dataframe(test_data_name, return_separate_X_and_y=True)

    description_data = pd.concat([train_x, test_x], axis=0)
    target_data = np.hstack([train_y, test_y])


    hc = KNeighborsTimeSeriesClassifier(n_jobs=-1)

    five_fold = StratifiedKFold(n_splits=5)

    X = description_data
    y = target_data

    print(X.shape, y.shape)

    cnt = 0
    l_dfs = []
    for train_index, test_index in five_fold.split(X, y):
    #         print("TRAIN:", train_index, "TEST:", test_index)
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y[train_index], y[test_index]

        print("-------"*10)    
        print("Started training fold {}".format(cnt+1))
        hc.fit(X_train, y_train)
        print("-------"*10)    
        print("Started prediction")
        preds = hc.predict(X_test)
        cnt+=1
        fold_df = pd.DataFrame(np.vstack([test_index, y_test, preds])).T
        fold_df.columns = ["index", "true_value", "prediction"]
        
        l_dfs.append(fold_df)
    
    pd.concat(l_dfs).to_csv("./results/"+dataset_name+".csv", index=False)    
    print("$$$$$$$"*10)
    
```

```python
def filter_data(x, dataset_names):
    if x in dataset_names:
        return True
    else:
        return False
```

```python
path = "./Meta_learning_Time_series_Classification/original_data/Univariate2018_ts/Univariate_ts/"
datasets = os.listdir(path)
dataset_names = pd.read_csv("datasets.txt", sep="\t").values
```

```python
from pathlib import Path
dsize = {}
for dataset_name in os.listdir(path):
    root_directory = Path(path+dataset_name+"/")
    dsize[dataset_name] = sum(f.stat().st_size for f in root_directory.glob('**/*') if f.is_file())




data = pd.DataFrame(dsize, index=[0]).T
data = data.reset_index()
data.columns = ["dataset_name", "byte_size"]
data = data.sort_values(["byte_size"])
data["to_drop"] = data.dataset_name.apply(lambda x: filter_data(x, dataset_names))
data = data[data.to_drop==True]
data = data.drop(["to_drop"], axis=1)
```
