```python
run1_1000 = pd.read_csv("1000_run1.csv")
run2_1000 = pd.read_csv("1000_run2.csv")
run3_1000 = pd.read_csv("1000_run3.csv")

run123_1000 = (run1_1000 + run2_1000 + run3_1000)/3
```
